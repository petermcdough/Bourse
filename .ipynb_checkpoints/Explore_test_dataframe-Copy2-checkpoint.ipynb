```python
from enum import Enum         # handles enumerations
import quandl                 # handles financial data imports
import pandas as pd           # handles dataframes
import sqlite3                # handles databases
import numpy as np            # handles maths computations
import os                     # handles directory browsing

from functools import partial # creates sub-functions of a chosen function
```

```python
# Classes
""" These are all classes used in the program """
class Data:
    """ This objects is a table, defined by:
        - the database name that contains the table (SQL)
        - the table name (SQL)
        - the data it contains as a dataframe
        When __init__ is called, it creates the table if it doesn't exist.
        The table can be created with columns name specified by the user, it not it creates a faint empty table
        (containing only one column named after config.DEFAULT_INDEX)"""
    
    def __init__(self, db_name, tab_name, columns = None):
        self.db_name = db_name
        self.tab_name = tab_name
        self._create(columns)
        self.data = self.get_data()
    
    def _connect(self):
        # establishes connection to the database and returns the connector
        # if the database doesn't exist yet, it creates it and the folder where it should be in
        if not os.path.exists(config.DB_FOLDER): # if the path where the database should be doesn't exist, create it
            os.mkdir(config.DB_FOLDER)
        os.chdir(config.DB_FOLDER) # choose folder where the database is located
        conn = sqlite3.connect(self.db_name)
        return conn
    
    def _query(self, query, dtype = None, reshape = True):
        # performs a query on self database
        # if dtype is 'df' returns the formatted data contained in the table as a dataframe
        # note that all columns containing 'DATE' are formatted to be compatible with pandas datetime format
        conn = self._connect()
        cur = conn.cursor()
        if dtype is None:
            answer = cur.execute(query)
            self._save_and_close(conn, cur)
        elif dtype == 'df':
            answer = pd.read_sql_query(query, conn) # fetch data
            for column in list(answer.columns[['DATE' in column for column in list(answer.columns)]]):
                answer[column] = pd.to_datetime(answer[column]) # formatting
            self._save_and_close(conn, cur)
        return answer
    
    def _save_and_close(self, conn, cur = None):
        # saves changes to the database and closes the connection to the database
        conn.commit()
        if cur is not None: # if a cursor has been created, close it too
            cur.close()
        conn.close()
        
    def _create(self, columns):
        # creates an empty table in a database if it doesn't exist
        # note that the table actually has one empty column DATE, since sqlite3 doesn't support empty tables
        if columns is None:
            query = "CREATE TABLE IF NOT EXISTS " + self.tab_name + " (" + config.DEFAULT_INDEX + " TIMESTAMP)"
        else:
            query = "CREATE TABLE IF NOT EXISTS " + self.tab_name + " " + columns
        self._query(query)

    def get_data(self):
        # returns the dataframe made from self table content
        query = "SELECT * FROM " + self.tab_name
        return self._query(query, dtype = 'df')
        
    def _save(self):
        # saves self.data in self.table
        conn = self._connect()
        self.data.to_sql(self.tab_name, conn, if_exists = 'replace', index=False)
        self._save_and_close(conn)
    
    def _merge(self, df):
        # replaces the content of self table with the union of self and df
        self.data = pd.merge(self.data, df, how='outer', sort=True)
    
    def _contains(self, column):
        # returns True if self data contains the column column
        return column in list(self.data)
    
    def _add_column(self, column_name):
        # adds a new column to self.data
        self.data[column_name] = None
    
    def _add_row(self, new_row):
        # adds the new row new_row to self.data
        self.data.loc[len(self.data)] = new_row

class Stock(Data, Enum):
    """ This object is a Data containing the stock data of a given company
        It is defined by :
        - a Data which it belongs to
        - a code, which is the code referred to by quandl (type : EURONEXT/ORA)
        - a last update date which is the last known date when the data as been refreshed
        Note that this object also has a useful attribute 'name', inherited from Enum """
    
    ORANGE = 'ORA'
    AIRBUS = 'AIR'

    def __init__(self, code):
        super().__init__(config.DB_STOCK_NAME, self.name) # creates empty table if it doesn't exist already
        self.code = config.MARKET + code
        self._set_last_refresh()
    
    def refresh(self):
        # updates stock data with the latest data from quandl, computes indicators for the new values and computes orders for all strategies in Strategy
        df = self._get_data_from_api()
        self._merge(df)
        self._compute_indicators()
        self._save()
        self._apply_strategies()
        self._set_last_refresh()
        print(self.name + " successfully updated!")
    
    def _set_last_refresh(self):
        # sets the last_refresh date and row of the stock
        # if the stock has just been created, last_refresh is None and last_row is -1
        if self.data.empty:
            self.last_refresh = None
            self.last_row = -1
        else:
            self.last_refresh = self.data.iloc[-1][config.DEFAULT_INDEX]
            self.last_row = self.data.iloc[-1].index
        
    def _get_data_from_api(self):
        # returns dataframe from quandl api and formats it
        df = quandl.get(self.code, start_date=config.START_DATE)
        df = df.reset_index()
        df = df.rename(index=str, columns={
                'Date':'DATE',
                'Open':'OPEN',
                'High':'HIGH',
                'Low':'LOW',
                'Last':'CLOSE',
                'Volume':'VOLUME',
                'Turnover':'TURNOVER'
        })
        return df

    def _compute_indicators(self):
        # updates self.data with computed indicators (for all known indicators)
        for indicator in Indicator:
            indicator.apply_on_stock(self)
            
    def _apply_strategies(self):
        # applies all known strategies to the stock
        for strategy in Strategy:
            strategy.apply_on_stock(self)

class Functions(Enum):
    """ This object contains all known functions to compute indicators, it cannot be initialised """
    
    def SMA(df, row, time_period, input = 'CLOSE'):
        # returns the Simple Moving Average value on provided df and row
        if row < time_period - 1:
            return None
        else: 
            return df[input][row-(time_period - 1):row+1].sum()/time_period

    def market(df, row, output):
        return df.at[row+1, output]

class Indicator(Enum):
    """ This object contains all known technical indicators.
        It is defined by a function capable of computing the indicator on a given dataframe"""
    
    SMA20 = partial(Functions.SMA, time_period = 20)
    SMA50 = partial(Functions.SMA, time_period = 50)
    SMA100 = partial(Functions.SMA, time_period = 100)
    
    def __init__(self, compute):
        self.compute = compute
    
    def apply_on_stock(self, stock):
        # updates stock.data with computed value of indicator
        if not stock._contains(self.name): # if the indicator is not in stock data yet, creates it
            stock._add_column(self.name)
            first_row_to_update = 0
        else:
            first_row_to_update = stock.last_row + 1
        if first_row_to_update < len(stock.data): # computes indicator for all rows after the stock last_refresh
            for row in range(first_row_to_update, len(stock.data)):
                stock.data.at[row, self.name] = self.compute(stock.data, row)
                
    def is_greater_than(self, df, indicator):
        # returns True if the self indicator is greater than the indicator of a df
        # note that df must have just one row
        if (df[indicator.name] is None) or (df[self.name] is None):
            return False
        else:
            return (df[indicator.name] < df[self.name]).bool()
    
    
class Strategy(Data, Enum):
    """ This object is a Data structure containing all current and past positions and performance on stocks
        It is defined by :
        - a Data structure (with the name of the strategy)
        - a frame, which is the frame of the dataframe the strategy analyzes to make a buy / sell decision
        - a condition to buy function that returns True if the stock should be bought
        - a condition to sell function that returns True if the stock should be sold
        - a buying_model function that sets the buying order strategy
        - a selling_model function that sets the selling order strategy
        - an is_new parameter that must be True if the strategy has just been created
        (otherwise the data will not update entirely)"""
    
    SMA50_SUPERIOR_TO_SMA20 = (
        partial(Indicator.SMA20.is_greater_than, indicator = Indicator.SMA50),
        partial(Indicator.SMA50.is_greater_than, indicator = Indicator.SMA20),
        True
    )
    
    def __init__(self, condition_to_buy, condition_to_sell, is_new = False, buying_model = partial(Functions.market, output='OPEN'), selling_model = partial(Functions.market, output='OPEN'), frame = lambda df,i: df.loc[[i]]):
        super().__init__(config.DB_PERFORMANCE_NAME, self.name, "( \
            OPENING_DATE TIMESTAMP, \
            CLOSING_DATE TIMESTAMP, \
            STOCK TEXT,             \
            STATUS TEXT,            \
            BUYING_PRICE REAL,      \
            LAST_PRICE REAL,        \
            PERFORMANCE REAL        \
        )")
        self.condition_to_buy = condition_to_buy
        self.condition_to_sell = condition_to_sell
        self.frame = frame
        self.buying_model = buying_model
        self.selling_model = selling_model
        self.is_new = is_new
        
    def apply_on_stock(self, stock):
        # updates self.data with the positions that were taken with the strategy on stock
        if self.is_new and self.data[self.data.STOCK == stock.name].empty: # if the strategy is new, we must take all stock values into account, there's a security in case one forgets to pass the is_new to False after the strategy has been used
            first_row_to_update = 0
        else:
            first_row_to_update = stock.last_row + 1
        
        if self._get_pending_position(stock) is None: # if the stock is not in current portfolio
            last_action = self.wait
        else:
            last_action = self.hold
        
        if first_row_to_update < len(stock.data):
            for row in range(first_row_to_update, len(stock.data)):
                focus = self.frame(stock.data,row)
                if self.condition_to_buy(focus):
                    if last_action != self.buy and last_action != self.hold: # Buy
                        action = self.buy
                    else: # Hold
                        action = self.hold
                elif self.condition_to_sell(focus):
                    if last_action == self.buy or last_action == self.hold: # Sell
                        action = self.sell
                    else: # Wait
                        action = self.wait
                else:
                    if last_action == self.buy: # Hold
                        action = self.hold
                    elif last_action == self.sell: # Wait
                        action = self.wait
                    else:
                        action = last_action
                action(row, stock)
                last_action = action
            self._save()        
    
    def buy(self, row, stock):
        # updates self.data after stock has been bought on a specified date index (row)
        position = Position(stock.data.at[row, 'DATE'], stock, self.buying_model(stock.data, row))
        self._update_position(position, len(self.data))
    
    def hold(self, row, stock):
        # updates self.data after stock is hold on a specified date index (row)
        position, index = self._get_pending_position(stock)
        position.update(stock.data.at[row, 'CLOSE'])
        self._update_position(position, index)
    
    def sell(self, row, stock):
        # updates self.data after stock has been sold on a specified date index (row)
        position, index = self._get_pending_position(stock)
        position.close(stock.data.at[row, 'DATE'], self.selling_model(stock.data, row))
        self._update_position(position, index)
    
    def wait(self, row, stock):
        # does nothing since wait means don't buy when you don't have stock
        pass
        
    def _update_position(self, position, index):
        # replaces (or creates) a position at specified index with position
        self.data.at[index] = [
            position.opening_date,
            position.closing_date,
            position.stock.name,
            position.status,
            position.buying_price,
            position.last_price,
            position.performance]
        
    def _get_pending_position(self, stock):
        # returns the pending position and index of stock or None if there isn't one
        index = self.data[(self.data.STATUS == 'pending') & (self.data.STOCK == stock.name)].index
        if len(list(index)) > 0 :
            row = np.asscalar(index)
            return (Position(
                opening_date= self.data.at[row, 'OPENING_DATE'],
                closing_date= self.data.at[row, 'CLOSING_DATE'],
                stock= Stock[self.data.at[row, 'STOCK']],
                status= self.data.at[row, 'STATUS'],
                buying_price= self.data.at[row, 'BUYING_PRICE'],
                last_price= self.data.at[row, 'LAST_PRICE']), row)
        else:
            return None
        
class Position:
    """ This object defines a position taken in the market
        It is defined by :
        - an opening_date which is the date the stock was bought
        - a closing_data which is the date the stock was sold
        - a stock
        - a status which is "pending" if the stock has been bought but not sold and "closed" when it has been bought and sold
        - a buying_price which is the price to which the stock wa bought
        - a last_price which is the last known price is a stock is owned or the sold price if the stock was sold
        - a performance which is the ratio between the last price and the buying_price (taking taxes into account)"""
    
    def __init__(self, opening_date, stock, buying_price, last_price = None, status = "pending", closing_date = None):
        self.opening_date = opening_date
        self.closing_date = closing_date
        self.stock = stock
        self.status = status
        self.buying_price = buying_price
        if last_price is None: # if no last_price specified that means the stock has just been bought
            self.last_price = buying_price
        else:
            self.last_price = last_price
        self._calc_performance()
    
    def update(self, new_price):
        # updates the stock last price and performance with the new_price
        self.last_price = new_price
        self._calc_performance()
    
    def close(self, closing_date, closing_price):
        # sells a stock
        self.status = "closed"
        self.closing_date = closing_date
        self.last_price = closing_price
        self._calc_performance()
        
    def _calc_performance(self):
        # returns the performance of a position including taxes
        # note that even if the position is not sold, it takes solding taxes into account
        self.performance = self.last_price / self.buying_price * (1 - config.TAX) * (1 - config.TAX)
```
