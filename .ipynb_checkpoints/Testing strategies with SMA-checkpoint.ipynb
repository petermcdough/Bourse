```python
def display_on_graphs(focus, x, y, figsize = (20,100)):
    fig, axis = plt.subplots(nrows = int(len(focus) / 4) + 1, ncols=4, sharex = False, sharey = False, figsize=figsize)
    plt.subplots_adjust(wspace = 0.05, hspace = 0.07)
    j = 0
    for ax in axis.reshape(-1):
        if j < len(focus):
            stock = focus[j]
            ax.plot(stock.data[x], stock.data[y])
            ax.set_title(stock.name)
            ax.tick_params(axis='both', which='both', bottom=False, labelbottom=False, left=False, labelleft=True)
            j = j+1
    plt.show()

def display_strategies_on_graphs(stocks, strategy, figsize = (20,130)):
    j = 0
    fig, axis = plt.subplots(nrows = int(len(stocks) / 4) + 1, ncols=4, sharex = False, sharey = False, figsize=figsize)
    for ax in axis.reshape(-1):
        if j < len(stocks):
            stock = stocks[j]
            ax.plot(stock.data.DATE, stock.data.CLOSE)
            ax.set_title(stock.name, fontsize = 8, y = 0.98)
            df = strategy.trades(stock = stock)
            for i in range(0, len(df)):
                opening_date = df.iat[i, 0] # index 0 for column is 'OPENING_DATE'
                ax.axvline(opening_date, color='grey', linestyle ='dashed', linewidth=0.5)
                opening_index = list(stock.data[stock.data.DATE == opening_date].index)[0]
                closing_date = df.iat[i, 1]
                if str(closing_date) == 'NaT':
                    closing_index = stock.last_row
                else:
                    closing_index = list(stock.data[stock.data.DATE == closing_date].index)[0]
                    ax.axvline(closing_date, color='grey', linestyle ='dashed', linewidth=0.5)                    
                sub_stock = stock.data.iloc[opening_index:(closing_index+1), :]
                if df.iat[i, -1] >= 1:
                    color = 'green'
                else:
                    color = 'red'
                ax.plot(sub_stock.DATE, sub_stock.CLOSE, color=color)
            #ax.set_xlim(left=stock.first_date, right=stock.last_refresh)
        ax.tick_params(axis='both', which='both', bottom=False, labelbottom=False, left=False, labelleft=False)
        j = j + 1
    plt.show()
```
