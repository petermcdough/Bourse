# Exploration functions

```python
def buy_condition(stock, row):
    date = None
    price = None
    status = None
    if np.isnan(row.SMA80) or np.isnan(row.SMA150):
        state = True
    else:
        state = row.SMA80 > row.SMA150
        if state == True:
            pos = stock.data.index.get_loc(row.Index)
            if pos + 1 == len(stock.data):
                price = None
                status = 'awaiting buy'
                date = row.Index
            else:
                next_row = stock.data.iloc[pos + 1]
                price = next_row.OPEN
                status = 'pending'
                date = next_row.name
    return {'state': state, 'date': date, 'price': price, 'status': status}

def sell_condition(stock, row, trade):
    date = None
    price = None
    status = None
    state = row.SMA80 < row.SMA150
    if state == True:
        pos = stock.data.index.get_loc(row.Index)
        if pos + 1 == len(stock.data):
            price = None
            status = 'awaiting sell'
            date = row.Index
        else:
            next_row = stock.data.iloc[pos + 1]
            price = next_row.OPEN
            status = 'closed'
            date = next_row.name
    return {'state': state, 'date': date, 'price': price, 'status': status}
```

# Testing SMA strategies

```python
stocks = load_stocks()
```

```python
SMA = File('SMA', 'Function').load()
```

```python
SMA_windows = [20, 50, 80, 100, 150, 200]
```

```python
SMA_indicators = []
for window in SMA_windows:
    SMA_indicators.append(Indicator('SMA' + str(window), compute = partial(SMA.execute, window = window)))
```

```python
for stock in stocks:
    for indicator in SMA_indicators:
        stock.add_indicator(indicator)
```

```python
def SMA_buy_condition(stock, row, SMA_1, SMA_2):
    value_1 = getattr(row, SMA_1.name)
    value_2 = getattr(row, SMA_2.name)
    if str(value_1) == 'nan' or str(value_2) == 'nan':
        return True
    else:
        return value_1 > value_2

def SMA_sell_condition(stock, row, trade, SMA_1, SMA_2):
    value_1 = getattr(row, SMA_1.name)
    value_2 = getattr(row, SMA_2.name)
    return value_2 > value_1
```

```python
strategies = []
for (SMA_1, SMA_2) in product(SMA_indicators, SMA_indicators):
    window_1 = SMA_1.compute.keywords['window']
    window_2 = SMA_2.compute.keywords['window']
    if window_1 < window_2:
        strategies.append(Strategy(
            name = SMA_1.name + "_" + SMA_2.name,
            buy_condition = partial(SMA_buy_condition, SMA_1 = SMA_1, SMA_2 = SMA_2),
            sell_condition = partial(SMA_sell_condition, SMA_1 = SMA_1, SMA_2 = SMA_2)))
```

```python
for strategy in strategies:
    strategy.refresh(stocks)
```

```python
def buy_condition(stock, row):
    if np.isnan(row.SMA80) or np.isnan(row.SMA150):
        return True
    else:
        return row.SMA80 > row.SMA150

def sell_condition(stock, row, trade):
    trade.update_price(row.CLOSE)
    condition_1 = (trade.performance < 0.80)
    condition_2 = row.SMA80 < row.SMA150
    return condition_1 or condition_2
```
